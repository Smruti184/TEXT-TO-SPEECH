# TEXT TO SPEECH USING PYTHON

```python
import tkinter as ttk
from tkinter import *
from tkinter import filedialog
from tkinter.ttk import Combobox
from tkinter import messagebox
import os
from playsound import playsound
root=Tk()
root.title("Text to speech converter")
root.geometry("1000x580+200+80")
root.resizable(False,False)
root.configure(bg="#F7AC40")
#root.mainloop()
from gtts import gTTS
import pyttsx3
tts=pyttsx3.init()

s_logo = PhotoImage(file = "D:\Downloads\s logo (1).png")
Label(root, image=s_logo,width=100,bg="#F7AC40").place(x=900,y=490)

def speak_now():
    text=text_box.get(1.0,END)
    if text:
        gender=gender_box.get()
        speed=speed_box.get()
        voices=tts.getProperty('voices')

        if(gender=='Male'):
            tts.setProperty('voice',voices[0].id)
          
        else:
            tts.setProperty('voice',voices[1].id)
            
            if(speed=='Fast'):
                tts.setProperty('rate',250)
                
            elif(speed=='Medium'):
                tts.setProperty('rate',150)
            else:
               tts.setProperty('rate',60)
        tts.say(text)
        tts.runAndWait()

logo_image=PhotoImage(file="D:\Downloads\P1.png")
root.iconphoto(False,logo_image)
#root.mainloop()
upper_frame=Frame(root,bg="#14A7DD",width=1200,height=160)
upper_frame.place(x=0,y=0)
                 
picture=PhotoImage(file="D:\Downloads\P2.png")
Label(upper_frame,image=picture,bg="#14A7DD").place(x=90,y=4)

#root.mainloop()
Label(upper_frame,text="Text to Speech Converter",font="TimesNewRoman 40 bold",bg="#14A7DD",fg='white').place(x=250,y=35)
#root.mainloop()
text_box=Text(root,font="calibri 20",bg="white",relief=GROOVE,wrap=WORD,bd=0)
text_box.place(x=30,y=180,width=940,height=180)
#root.mainloop()

gender_box=Combobox(root,values=['Male','Female'],font="Robote 12",state='r',width=12)
gender_box.place(x=340,y=400)
gender_box.set('Male')
#root.mainloop()

speed_box=Combobox(root,values=['Fast','Medium','Slow'],font="Robote 12",state='r',width=12)
speed_box.place(x=540,y=400)
speed_box.set('Medium')

Label(root,text="Select Voice",font='TimesNewRoman 15 bold',bg="#F7AC40",fg="white").place(x=342,y=370)
Label(root,text="Select Speed",font='TimesNewRoman 15 bold',bg="#F7AC40",fg="white").place(x=540,y=370)
#root.mainloop()

play_button=PhotoImage(file="D:\Downloads\play (2).png")
play_btn=Button(root,compound=LEFT,image=play_button,bg='white',width=53,height=57,font="arial 14 bold",borderwidth='0.1c',command=speak_now)
play_btn.place(x=472,y=440)
root.mainloop()
```
